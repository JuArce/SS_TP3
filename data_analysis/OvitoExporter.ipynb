```python
import numpy as np
import pandas as pd
```

```python
# Read output files csv
N = [10,50,100,200]
slot = [0.01,0.02,0.05]
output_file = "output"
static_file = "static"
end_file = ".csv"
output_files = []
static_files = []
for n in N:
    for s in slot:
        output_files.append(pd.read_csv(f"../src/main/resources/output/parameters/{output_file}_N_{n}_slot_{s}{end_file}"))
        static_files.append(pd.read_csv(f"../src/main/resources/input/parameters/{static_file}_N_{n}_slot_{s}{end_file}", sep='\t', header=None))
assert len(output_files) == len(N) * len(slot)
assert len(static_files) == len(N) * len(slot)
```

```python
# Fix static data
aux = []
for s_data in static_files:
    s_data = s_data[4:]
    s_data[['radius','mass']] = s_data[0].str.split(',',expand=True)
    s_data = s_data.drop(columns=0)
    aux.append(s_data)
static_files = aux
```

```python
static_files[0]
```

```python
#Fixing index on static data
aux = []
for s_data in static_files:
    s_data.index = np.arange(1, len(s_data) + 1)
    aux.append(s_data)
static_files = aux
```

```python
aux = len(output_files[0])
max_iteration = int(output_files[0].loc[aux-1]['iteration'])
max_id = int(output_files[0].loc[aux-1]['id'])
```

```python
rows2drop = list(range(1,int(max_iteration)+1))
rows2drop
```

```python
rows2drop = [x*11 for x in rows2drop]
rows2drop
```

```python
#Fixing output data, removes event calls
aux=[]
for output in output_files:
    aux.append(output.drop(rows2drop))
output_files = aux
output_files
```

```python
join_data = []
for i in range(0, len(static_files)):
    join_data.append(output_files[i].join(static_files[i], on='id'))
assert len(join_data) == len(static_files)
assert len(join_data) == len(output_files)
print(join_data)
```

```python
#Dropping useless data
aux = []
for data in join_data:
    aux.append(data.drop(columns=['iteration']))
join_data = aux
print(join_data)
```

```python
start_line = [max_id,np.nan,np.nan,np.nan,np.nan,np.nan,'\n']
```

```python
aux = []
for data in join_data:
    data = data.dropna()
    data = data.astype({'id':int,'mass': str})
    aux.append(data)
    print(data)
join_data = aux
```

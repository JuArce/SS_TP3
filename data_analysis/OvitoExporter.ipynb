```python
import numpy as np
import pandas as pd
```

```python
# Read output csv
file = "_N_100_L_10_n_0.1.csv"
data = pd.read_csv(f"../src/main/resources/output/parameters/output{file}")
print(data)
```

```python
#Read static data
s_data = pd.read_csv(f"../src/main/resources/input/parameters/static{file}")
s_data.columns = ['radius']
print(s_data.columns)
```

```python
#Saving max data
max_id = data['id'].max()
max_iteration = data['iteration'].max()
print(f'max_id: {max_id}\nmax_iteration: {max_iteration}')
```

```python
join_data = data.join(s_data,on='id')
print(join_data)
```

```python
#Dropping useless data
join_data = join_data.drop(columns=['iteration'])
print(join_data)
```

```python
start_line = [max_id,np.nan,np.nan,np.nan,np.nan,'\n']
```

```python
a = (np.arange(max_iteration+1) * max_id) - 1
a = np.delete(a,0)
a
```

```python
join_data = join_data.astype({'id':int,'radius': str})
join_data
```

```python
import csv
join_data = join_data.sort_index().reset_index(drop=True)
aux_i = 0
aux_data = pd.DataFrame(columns=join_data.columns,index=range(0,max_id+1))
aux_data
```

```python
#empty file
try:
    f = open(f"./resources/ovito_exporter_{file}", "w+")
    f.close()
except:
    print("file doens't exist")
```

```python
empty_df = pd.DataFrame(columns=join_data.columns,index=[0])
empty_df.loc[0] = start_line
for x in a:
    aux_data = join_data.loc[aux_i:x]
    print(aux_data)
    aux_i = x + 1
    empty_df.to_csv(f"./resources/ovito_exporter_{file}",sep=' ', index=False, header=False,quoting=csv.QUOTE_NONE,escapechar=" ", mode='a')
    aux_data.to_csv(f"./resources/ovito_exporter_{file}",sep=' ', index=False, header=False,quoting=csv.QUOTE_NONE,escapechar=" ", mode='a')
```
